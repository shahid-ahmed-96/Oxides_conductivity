# Code for data extraction of Oxides

#### Disclaimer : This code when written was working fine uptil March 2023, but due to some database relocation (Duke AFLOW database) this code is no longer working.   

```python
from matminer.data_retrieval.retrieve_AFLOW import AFLOWDataRetrieval
```

```python
adr=AFLOWDataRetrieval()
```

```python
df = adr.get_dataframe(criteria={"species": "O"}, properties=["agl_thermal_conductivity_300K", "natoms","agl_bulk_modulus_static_300K","ael_poisson_ratio", "agl_debye", "agl_gruneisen", "agl_heat_capacity_Cp_300K","agl_thermal_expansion_300K", "density", "Egap", "Bravais_lattice_lattice_type", "ael_speed_sound_average"],index_auid=False)
```

```python
df2=df
df2=df2.set_index("compound")
df2=df2[["agl_thermal_conductivity_300K", "natoms","agl_bulk_modulus_static_300K","ael_poisson_ratio", "agl_debye", "agl_gruneisen", "agl_heat_capacity_Cp_300K","agl_thermal_expansion_300K", "density", "Egap", "Bravais_lattice_lattice_type", "ael_speed_sound_average"]]
df2 = df2.dropna()
```

```python
final_dataset=df2[~df2.index.duplicated()]
final_dataset
```

### Storing the dataset to a csv file named Dataset

```python
import pandas as pd
dataset=pd.DataFrame(final_dataset)
dataset.to_csv('Dataset.csv')
```
